```python
import pandas as pd
```

```python
def creat_df_month(nmonth,year):
    frames=[]
    for i in range(1,nmonth+1,1):
        name = 's_d_t_0' + str(i) + '_' + str(year) + '.csv'
        df_tmp = pd.read_csv(name, encoding = "windows-1250", header = None)
        frames.append(df_tmp)
    df = pd.concat(frames) 
    return df
```

```python
def creat_df_year(number,years):
    frames=[]
    for year in years:
                name = 's_d_t_' + str(number) + '_' + str(year) + '.csv'
                df_tmp = pd.read_csv(name, encoding = "windows-1250", header = None)
                frames.append(df_tmp)
    df = pd.concat(frames)
    return df
```

```python
def df_wrag():
    df = pd.concat([creat_df_month(3,2019),creat_df_year(375,[2017,2018])])
    new_df = df[df[0]== 352200375]
    new_df = new_df.drop([0,1,5,6,8,10,11,12,14,16,17,18,20,21,22], axis=1)
    new_df.columns=['Year','Month','Day','Wind','Temp','Humidity','Press','Rain']
    new_df = new_df.sort_values(by=['Year','Month','Day'])
    new_df = new_df.reset_index()
    new_df = new_df.drop(['index'], axis=1)
    return new_df
df = df_wrag()
df
```

```python
df.to_csv("weather_data.csv",encoding='utf-8',index=False)
```
